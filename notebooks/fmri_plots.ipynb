```python
#Make mask centered around [-43, -4, 14]
import nipype.interfaces.fsl as fsl

data_loc = os.environ["DATA_LOC"]
server_scripts = os.environ["SERVER_SCRIPTS"]
out_path = os.path.join(data_loc, 'derivatives', 'rois')
anatfile = os.path.join(out_path, 'tpl-MNI152NLin2009cAsym_res-02_desc-brain_T1w','tpl-MNI152NLin2009cAsym_res-02_desc-brain_T1w.nii.gz')
space = os.path.basename(anatfile)
space = os.path.splitext(os.path.splitext(space)[0])[0]
T = nib.load(anatfile).affine

cur_row = pd.DataFrame(data={'Name': 'Fig_4_9', 'x': [-43], 'y': [-4], 'z': [14]})
vox_cor = mni2vox([cur_row.iloc[0]['x'], cur_row.iloc[0]['y'], cur_row.iloc[0]['z']], T)
if not os.path.exists(os.path.join(out_path, space)):
    os.makedirs(os.path.join(out_path, space))

maths1 = fsl.ImageMaths()
maths1.inputs.in_file = anatfile
maths1.inputs.op_string = '-mul 0 -add 1 -roi %s 1 %s 1 %s 1 0 1'%(vox_cor[0], vox_cor[1], vox_cor[2])
maths1.inputs.out_data_type = 'float'
maths1.inputs.out_file = os.path.join(out_path, space, cur_row.iloc[0]['Name']+'_point.nii.gz')
maths1.run()

maths2 = fsl.ImageMaths()
maths2.inputs.in_file = os.path.join(out_path, space, cur_row.iloc[0]['Name']+'_point.nii.gz')
maths2.inputs.op_string = '-kernel sphere 5 -fmean'
maths2.inputs.out_data_type = 'float'
maths2.inputs.out_file = os.path.join(out_path, space, cur_row.iloc[0]['Name']+'_sphere.nii.gz')
maths2.run()

maths3 = fsl.ImageMaths()
maths3.inputs.in_file = os.path.join(out_path, space, cur_row.iloc[0]['Name']+'_sphere.nii.gz')
maths3.inputs.op_string = '-bin'
maths3.inputs.out_file = os.path.join(out_path, space, cur_row.iloc[0]['Name']+'_bin.nii.gz')
maths3.run()

print("***********************************************")
print("ROI saved in %s"%(os.path.join(out_path, space, cur_row.iloc[0]['Name']+'_bin.nii.gz')))
print("***********************************************")
```
